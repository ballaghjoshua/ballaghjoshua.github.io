## Part II: Income Data
In the rest of this notebook, we will be loading additional datasets against which we will be comparing our school data. First up is data on income and unemployment by parish in 2018, taken from the USDA. The original dataset can be found here: https://data.ers.usda.gov/reports.aspx?ID=17828

```python
employment_df = pd.read_excel('UnemploymentReport.xlsx')
# For some reason, this file doesn't like to be read as an Excel file (despite that being what the Census distributes it as).
# Reading it as a CSV seems to work, though!
employment_df

# The year columns are unemployment rates as a percent.
# The first row contains the numbers for the state as a whole; each row after that observes a single parish.
```

Now that we've loaded in the table, let's clean it up a bit.

```python
employment_df.columns = employment_df.loc[0] # Assign columns proper names
employment_df = employment_df[1:66] # Cut off extra rows
```

```python
# To get rid of some annoying warnings that clutter the notebook:
pd.options.mode.chained_assignment = None  # default='warn'

employment_df.reset_index(drop = True, inplace = True) # Reset our index
employment_df['Median Household Income (2018)'] = employment_df['Median Household Income (2018)'].astype('float64') # set the type to float for doing math later

#Fix up the 'Name' column so that the strings are consistent with the school performance 2019 dataset
employment_df['Name'] = employment_df['Name'].str[:-4]
employment_df.at[0, 'Name'] = 'Louisiana'

employment_df
# Now let's look at our beautiful new table!
```

### SPS Scores vs. Median Household Income (2018)

Now let's see if there's a relationship between median household income and SPS.  

To do this easily, first we must integrate data from across the two tables. We will add a column to the 2019 School Performance dataframe titled "Parish Median Income" to represent Median Household Income for 2018 according to the USDA data.  

Much of this integration process is necessary because of the relationship between school systems and parishes in Louisiana. The set of all school systems != the set of all parishes. Every parish has a school system, but not every school system belongs to a parish!

```python
system_means_2018 = performance2019_df.groupby('School System')['2018 SPS'].mean() # Series of school system mean SPS for 2018
system_means_2019 = performance2019_df.groupby('School System')['2019 SPS'].mean() # Series of school system mean SPS for 2019
```

```python
# Make a list of parishes and a new dataframe
parishes = employment_df['Name'].unique() # List of Parish names in Louisiana
parishes_df = pd.DataFrame(parishes, columns = ['Name'])
parishes_df = parishes_df[1:] # Drop "Louisiana", as we only want parishes in this dataframe

# Construct series of data to add to our dataframe:
parish_median_income = employment_df['Median Household Income (2018)'][1:].tolist()
parish_means_2018 = []
parish_means_2019 = []

# Currently, De Soto and La Salle parishes are not showing up in our system means lists. 
# There may be a string matching error. We'll need to account for it!
for parish in parishes:
    if parish in system_means_2018:
        parish_means_2018.append(system_means_2018[parish])
    if parish in system_means_2019:
        parish_means_2019.append(system_means_2019[parish])
    if parish == 'De Soto Parish':
        parish_means_2018.append(system_means_2018['DeSoto Parish'])
        parish_means_2019.append(system_means_2019['DeSoto Parish'])
    if parish == 'La Salle Parish':
        parish_means_2018.append(system_means_2018['LaSalle Parish'])
        parish_means_2019.append(system_means_2019['LaSalle Parish'])


#plt.plot(df1.start_datetime, df1.speed.astype(float),'b+')

# Add series to parishes_df
parishes_df['Mean SPS 2018'] = parish_means_2018
parishes_df['Mean SPS 2019'] = parish_means_2019
parishes_df['Median Household Income 2018'] = parish_median_income

# Plot 2018 relationship!
sns.lmplot(x='Median Household Income 2018',y='Mean SPS 2018',data = parishes_df,fit_reg=True)
ax = plt.gca()
ax.set_title("Relationship Median Household Income and Mean School Performance Score \n (by Parish, 2018)")

parishes_df.corr() # display correlations between variables
```

#### Making sense of the relationship
There is a moderate correlation between Mean SPS and Median Household Income by parish, at about 52%. This relationship is not particularly strong, but it does exist.

We suspect that at a more granular level, the relationship between SPS and median household income would be stronger. In this case, one would need to look on a school-by-school basis, examining median household income of students enrolled at the school. However, this data is hard to come by, if it is available at all! 

This logically follows. In a given school system, there can exist both the "nice" schools located in wealthier neighborhoods and lower-scoring schools located in poorer neighborhoods. Examining the strength of this relationship in a given parish would make for a great case study in the future, but is currently beyond the scope of our project.

## Part III: Analyzing ACT Scores by Parish (2020)

Now we're going to look at ACT scores, focusing more on how they are broken down by parish, and whether or not the scores of parishes correlate with an inherent aspects of the parish. To do this, we will read in two more data frames from the LA Dept of Education. One is the ACT Scores of 2020 broken down by parish, and the other is student attributes data, also broken down by parish. We will combine these in order to have a data frame that contains information not only on student attributes, but also student scores, so that we can look at correlations. 

```python
# load up 2020 Parish ACT data
ParishACT20df = pd.read_excel('ACT-Scores---Class-of-2020.xlsx')

# clean 2020 Parish ACT data:

# first, we will rename the columns
ParishACT20df.columns = ['School System Code','School System','2019-2020 Student Count','2019/20 Average ACT']

# get rid of the first few rows, which are unnecessary to us
for x in range(0, 5):
    ParishACT20df = ParishACT20df.reindex(ParishACT20df.index.drop(0)).reset_index(drop=True)

# load up 2020 Student Attributes data
Students20Df = pd.read_excel('feb-2020-multi-stats-(total-by-site-and-school-system).xlsx')

# clean students attributes data:

# get rid of unnecessary rows
for x in range(0, 6):
    Students20Df = Students20Df.reindex(Students20Df.index.drop(0)).reset_index(drop=True)

# rename the columns
Students20Df.columns = ['School System','School System Name','Sites Reporting','Total Enrollment',\
                      '% Female','% Male','American Indian','Asian','Black','Hispanic','Hawaiian/Pacific Islander',\
                      'White','Multiple Races (Non-Hispanic)','Minority','% Fully English Proficient',\
                      '% Limited English Proficiency','Infants (Sp Ed)','Pre-School (Sp Ed)','Pre-K (Reg Ed)',\
                      'Kindegarten','Grade 1','Grade 2','Grade 3','Grade 4','Grade 5','Grade 6','Grade 7',\
                      'Grade 8','Grade 9','Grade T9','Grade 10','Grade 11','Grade 12','Extension Academy',\
                      '% Economically Disadvantaged']


# now we will merge StudentsDf and ParishACTdf into a singular dataset, merging on School System

# first, set index of StudentsDf to be School System
Students20Df = Students20Df.set_index('School System')
ParishACT20df = ParishACT20df.set_index('School System Code')

# Perform inner merge on index
Parishes2020 = pd.merge(ParishACT20df, Students20Df, left_index=True, right_index=True)

# display the data frame, which now possesses both student attributes and ACT scores (2020), broken down by parish. 
Parishes2020
```

```python
# Now that this data is all housed in a singular data frame, we can look to see if there is correlation
# First, we are going to investigate the relationship between ACT scores of a parish and the percent of economically disadvantaged students in that parish

# Cast ACT and % Economically Disadvantaged to Numeric
Parishes2020["2019/20 Average ACT"] = pd.to_numeric(Parishes2020["2019/20 Average ACT"])
Parishes2020['% Economically Disadvantaged'] = pd.to_numeric(Parishes2020['% Economically Disadvantaged'])


# using seaborn, we will plot this relationship

sns.lmplot(x='% Economically Disadvantaged',y='2019/20 Average ACT',data=Parishes2020,fit_reg=True)
ax = plt.gca()
ax.set_title("Relationship between ACT Scores and % of Economically Disadvantaged Students (by Parish)")

# look at correlation numerically
Parishes2020.corr()
```

From the graph above, we do in fact see that there is a correlation between the % of economically disadvantaged students that a parish has, and the overall average ACT score of that parish. The correlation is negative, meaning that as the % of economically disadvantaged students increases, the average ACT score decreases. However, although there is correlation, we can immediately infer causation. 

## Part IV: Analyzing ACT Scores by Subgroups (2012 - 2020)

We will also be looking at data on ACT scores as broken by subgroups (ethnicity, English learners, economically disadvantaged, etc), from the years 2012 to 2020. We will be drawing this data from the LA Department of Education as well. It comes in the form of three datasets: Subgroup ACT scores from 2020, Subgroup ACT scores from 2019, and Subgroup ACT scores from 2012 - 2018. Below you will see some code that reads in these three data sets. Analyzing these data frames as standing alone is useful (particularly the 2020 scores, as this gives an image of the current state of education), but we will also merge these tables to make a unified data frame in order to capture trends of particular subgroups and parishes over time. 

```python
## read in the 2020 Subgroup ACT data
SubACT20df = pd.read_excel('2020-state-district-act-subgroup-performance.xlsx')
SubACT20df.head()

# now we will tidy some of this up

# first, we will rename the columns
SubACT20df.columns = ['School System Code', 'School System Name (Parish)','Subgroup','ACT',
              '# Students Scoring 18+','# Students Scoring 21+']

# get rid of the first few rows, which are unnecessary to us
for x in range(0, 4):
    SubACT20df = SubACT20df.reindex(SubACT20df.index.drop(0)).reset_index(drop=True)

# now, I'll replace the '~' with NaNs
SubACT20df = SubACT20df.replace('~', np.nan)

# convert 2020 ACT scores to numeric
SubACT20df["ACT"] = pd.to_numeric(SubACT20df["ACT"])



# read in the 2019 Subgroup ACT data
SubACT19df = pd.read_excel('2019-state-district-act-subgroup-performance.xlsx')
SubACT19df.head()

# first, we will rename the columns
SubACT19df.columns = ['School System Code', 'School System Name (Parish)','Subgroup','ACT',
              '# Students Scoring 18+','# Students Scoring 21+']

# get rid of the first few rows, which are unnecessary to us
for x in range(0, 3):
    SubACT19df = SubACT19df.reindex(SubACT19df.index.drop(0)).reset_index(drop=True)
    
# now, I'll replace the '~' with NaNs
SubACT19df = SubACT19df.replace('~', np.nan)

# Now we have two dataframes with 2019 ACT Scores and 2020 ACT Scores. 
# Let's look at the 2020 ACT Dataframe to get an idea of what it looks like!
SubACT20df
```

```python
# With the 2020 dataframe read in, we can start performing some analysis of the current performance of particular subgroups

# create pivot table of 2020 Subgroup ACT data
SubACT20df.pivot_table(
    index="School System Name (Parish)", columns=["Subgroup"],
    values="ACT", aggfunc=np.mean
).plot.barh(figsize = (26,28))
```

In theory, this data visualization would give you an idea of how each subgroup within a parish is performing on the ACT. However, as you can see, it's quite crowded and hard to make sense of. Instead, let's focus on just a particular subgroup, and order the scores, to get more of a sense of what story this data can tell.

```python
# we will focus on the subgroup of Black/African American students

SubACT20df[SubACT20df.Subgroup == "Black or African American"]\
    .sort_values("ACT").\
    plot(x="School System Name (Parish)",y=["ACT"],kind="barh",figsize = (10, 25),\
    title="Avg ACT Score of Black or African American Students (2020)")
```

The graphic above gives a sense of how the subgroup of Black/African American students perform on the ACT, as broken down by parish. Because the graphic is order, we can see which parish has the highest subgroup score on the ACT, and which has the lowest. The scores themselves though do not tell a complete story. Do the schools that have the highest ACT scores of Black students also have the lowest disparity between Black students' scores and overall average scores? To find this out, we need to calculate disparity (difference between average ACT score of a parish and average ACT score of a particular subgroup within a parish). Unfortunately the 2020 Subgroup ACT Scores data frame does not already that data logged, so we will have to perform some operations on the data frame in order to find and log it. 

```python
# Here we will alter the 2020 Subgroup ACT Scores dataframe in order to log the disparity between each subgroup's score and the average score of their parish

# We will do this by extracting a parish's overall average score from the data frame, creating a dictionary that possesses parishes as keys and the overall average scores as their values. This dictionary will then be used fill a new column, and that new column will then be used computer the disparity value of any particular subgroup of any particular parish. 

# create new column
SubACT20df['Overall Avg Score'] = SubACT20df['School System Name (Parish)']

# create dictionary containing average scores, then use dictionary to fill new column of SubACT20df
new = SubACT20df[SubACT20df.Subgroup == "Total Population"][["School System Name (Parish)","ACT"]].copy()
dict_scores = dict(zip(new["School System Name (Parish)"], new["ACT"]))
dict_scores
SubACT20df.replace({"Overall Avg Score": dict_scores},inplace=True) 
SubACT20df["Overall Avg Score"] = pd.to_numeric(SubACT20df["Overall Avg Score"])

# using new column that was added, we can now create a column that computes the disparity between a subgroup's score and the average score of their parish. 
SubACT20df["Disparity"] = SubACT20df["ACT"] - SubACT20df["Overall Avg Score"]

# We'll call the dataframe, and now in the newly created last column "Disparity" you can see the change that has been
# For example, by looking at the data frame you can see that the statewide score of the Hispanic/Latinx subgroup is 1.1 less than the statewide score of the total population
SubACT20df
```

Now that we have this metric, we can focus on the question that we asked before. Do the schools that have the highest ACT scores of Black students also have the lowest disparity between Black students' scores and overall average scores? Let us plot the disparity broken down by parish to see. 

```python
# plot disparity between Black/AfricanAmerican subgroup ACT score to Total Population ACT score

SubACT20df[SubACT20df.Subgroup == "Black or African American"]\
    .sort_values("Disparity",ascending=False).\
    plot(x="School System Name (Parish)",y=["Disparity"],kind="barh",figsize = (10, 25),\
    title="Disparity between Parish's Avg ACT Score of All Students and Avg ACT Score of Black Students (2020)")
```

From this visualization of the disparity of ACT scores for the particular subgroup of Black/African American students, we are able to see a few things. When comparing this graphic with the graphic directly preceding, we can see that even though Zachary Community School District has the highest average ACT score of Black/African American students among parishes, it does not have the lowest level of disparity of scores. Instead, it falls in the middle. Meanwhile, the parishes with some of the lowest disparity (Tensas, St. Helena) don't necessarily have the highest average scores of Black students. In fact, they fall in the lowest third of parishes when numerically ordering the average ACT scores of Black students.
